```python
import matplotlib.dates as ndates
import matplotlib.ticker as ticker
from matplotlib import rcParams
from matplotlib.path import Path
from matplotlib.patches import PathPatch
```

```python
np.random.seed(42)
```

```python
# data preparation
dates = pd.date_range(start = '2024-02-01', periods = 100, freq = 'D')
initial_rate = 75
drift = 0.003
volatility = 0.1
returns = np.random.normal(drift, volatility, len(dates))
rates = initial_rate * np.cumprod(1 * returns)

x, y = dates, rates
```

```python
# general parameters
fig, ax = plt.subplots(figsize = (10, 6))
plt.title('Daily visitors', fontsize = 18, color = 'black')
rcParams['font.family'] = 'Dejavu Sans'
rcParams['font.size'] = 14

#axis Y to the right
ax.yaxis.tick_right()
ax.yaxis.set_label_position('right')
```

```python
# axis
x_interval = 21
ax.xaxis.set_major_formatter(ndates.DateFormatter('%d %b'))  # Corrected formatter
ax.xaxis.set_major_locator(ndates.DayLocator(interval=x_interval))  # Corrected locator
ax.yaxis.set_major_locator(ticker.MultipleLocator(50))  # Locator for Y-axis ticks
```

```python
# grid
ax.grid(color='gray', linestyle=(0, (10, 10)), linewidth = 0.5, alpha = 0.6)
ax.tick_params(axis = 'x', colors = 'black')
ax.tick_params(axis = 'y', left=False, labelleft = False)
```

```python
# borders
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['bottom'].set_color('black')
ax.spines['left'].set_color('white')
ax.spines['left'].set_linewidth(1)
```

```python
# remove ticks from axis Y
ax.tick_params(axis = 'y', length = 0)
```

```python
# add year to the first date on the axis
def custom_date_formatter(t, pos, dates, x_interval):
    date = dates[pos * x_interval]
    if pos == 0:
        return date.strftime("%d %b \%y")
    else:
        return date.strftime('%d %b')
ax.xaxis.set_major_formatter(ticker.FuncFormatter((lambda x, pos: custom_date_formatter(x, pos, dates=dates, x_interval=x_interval))))
```

```python
# tricks label color
[t.set_color('#808079') for t in ax.yaxis.get_ticklabels()]  # For y-axis tick labels
[t.set_color('#808079') for t in ax.xaxis.get_ticklabels()]  # For x-axis tick labels
```

```python
# gradient
numeric_x = np.array([i for i in range(len(x))])
numeric_x_patch = np.append(numeric_x, max(numeric_x))
numeric_x_patch = np.append(numeric_x_patch[0], numeric_x_patch)
y_patch = np.append(y, 0)
y_patch = np.append(0, y_patch)

path = Path(np.array([numeric_x_patch, y_patch]).transpose())
patch = PathPatch(path, facecolor='none')
plt.gca().add_patch(patch)

ax.imshow(numeric_x.reshape(len(numeric_x), 1),  interpolation="bicubic",
                cmap=plt.cm.Greens,
                origin='lower',
                alpha=0.3,
                extent=[min(numeric_x), max(numeric_x), min(y_patch), max(y_patch) * 1.2],
                aspect="auto", clip_path=patch, clip_on=True)
```

```python
# Plotting main line
y_chart = y_patch
y_chart[0] = y_chart[1]
y_chart[-1] = y_chart[-2]
ax.plot(numeric_x_patch, y_chart, color='#268358', linewidth=2)
```

```python
# fix a grey line of imshow
ax.plot([max(numeric_x_patch), max(numeric_x_patch)], [0, max(y)], color='white', linewidth=2)

plt.savefig('high_quality_plot.png', dpi=500, bbox_inches='tight')
plt.show();
```

```python
# Import the '../data/car-sales.csv' into a DataFame called car_sales and view
# Try to plot the 'Price' column using the plot() function
df = pd.read_csv('https://raw.githubusercontent.com/Rstam59/TaskDataRepoForStudents/refs/heads/main/car-sales.csv')
df['Price'] = df['Price'].str.replace('[\$,]', '', regex=True).astype(float)
df['Price'].plot()
plt.show()
```

```python
# Remove the symbols, the final two numbers from the 'Price' column and convert it to numbers
# df['Price'] = df['Price'].astype(str).str[:-2]
# df['Price'] = df['Price'].astype(float)
df['Price'] = df['Price'].astype(str).str.replace('[\$,]', '', regex=True)
df['Price'] = df['Price'].str[:-2]
df['Price'] = df['Price'].replace('', np.nan)
df['Price'] = df['Price'].astype(float)
```

```python
# Add a column called 'Total Sales' to car_sales which cumulatively adds the 'Price' column
# Add a column called 'Sale Date' which lists a series of successive dates starting from today (your today)
df['Total Sales'] = df['Price'].cumsum()
from datetime import datetime, timedelta
num_rows = len(df)
start_date = datetime.today()
date_list = [start_date + timedelta(days=i) for i in range(num_rows)]
df['Sale Date'] = date_list
print(df[['Price', 'Sale Date']].head())
plt.plot(df['Total Sales'])
plt.title('Cumulatively Total Sales')
plt.grid(True)
plt.show()
```

```python
print(df.head())
```

```python
# Use the plot() function to plot the 'Sale Date' column versus the 'Total Sales' column
plt.plot(df['Sale Date'], df['Total Sales'])
plt.title('Sale Date vs Total Sales')
plt.xlabel('Sale Date')
plt.ylabel('Total Sales')
plt.grid(True)
plt.show
```

```python
# Convert the 'Price' column to the integers
# Create a scatter plot of the 'Odometer (KM)' and 'Price' column using the plot() function
df['Price'] = df['Price'].astype(str).str.replace('[\$]', '', regex=True).astype(float).astype(int)
plt.scatter(df['Odometer (KM)'], df['Price'])
plt.title('Odometer (KM)')
plt.xlabel('Price')
plt.ylabel('Odometer (KM)')
plt.grid()
plt.show()
```

```python
# Replicate the above plot in whichever way you see fit
# Note: The method below is only one way of doing it, yours might be
# slightly different
# Create DataFrame with patients over 50 years old
# Create the plot
# Plot the data
# Customize the plot
# Add a meanline
over_50 = heart_disease[heart_disease['age'] > 50]
fig, ax = plt.subplots()
ax.scatter(over_50['age'], over_50['chol'])
ax.set(title='Cholesterol levels of patients over 50',
       xlabel='Age',
       ylabel='Cholesterol')
ax.axhline(over_50['chol'].mean(), linestyle='--', color='blue', label='Mean Cholesterol')
ax.legend()
plt.show()
```

```python
# Reproduce the same figure as above with the "seaborn-whitegrid" style
# Create the plot
# Plot the data
# Customize the plot
# Add a meanline
plt.style.use('seaborn-v0_8-whitegrid')
over_50 = heart_disease[heart_disease['age'] > 50]
plt.figure(figsize=(10, 6))
plt.scatter(over_50['age'], over_50['chol'], c='red', label='Age vs Cholesterol')
mean_chol = over_50['chol'].mean()
plt.axhline(mean_chol, linestyle='--', color='blue', label=f'Mean Cholesterol: {mean_chol:.0f}')
plt.title('Age vs Cholesterol (50 yaşdan yuxarı)', fontsize=14)
plt.xlabel('Age')
plt.ylabel('Cholesterol')
plt.legend()
plt.grid(True)
plt.show
```

```python
# Replot the same figure as above except change the "cmap" parameter
# of scatter() to "winter"
# Also change the "color" parameter of axhline() to "red"
# Create the plot
# Plot the data
# Customize the plot
# Add a meanline
plt.style.use('seaborn-v0_8-whitegrid')
over_50 = heart_disease[heart_disease['age'] > 50]
plt.figure(figsize=(10, 6))
scatter = plt.scatter(over_50['age'], over_50['chol'],
                      c=over_50['chol'], cmap='winter', label='Age vs Cholesterol')
plt.colorbar(scatter, label='Cholesterol')
mean_chol = over_50['chol'].mean()
plt.axhline(mean_chol, linestyle='--', color='red', label=f'Mean Cholesterol: {mean_chol:.0f}')
plt.title('Age vs Cholesterol (50 yaşdan yuxarı)', fontsize=14)
plt.xlabel('Age')
plt.ylabel('Cholesterol')
plt.legend()
plt.grid(True)
plt.show()
```
