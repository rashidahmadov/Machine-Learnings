```python
df.drop(['Id', 'STATE'], axis=1, inplace=True)
```

```python
X = df.drop('Risk_Flag', axis=1)
y = df['Risk_Flag'].copy()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)
```

```python
num_features = X_train.select_dtypes(include=[np.number]).columns
cat_features = X_train.select_dtypes(exclude=[np.number]).columns
```

```python
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.compose import ColumnTransformer
from sklearn.ensemble import RandomForestClassifier
```

```python
num_pipeline = Pipeline([
    ('impute', SimpleImputer(strategy='median')),
    ('scale', StandardScaler())
])

cat_pipeline = Pipeline([
    ('impute', SimpleImputer(strategy='most_frequent')),
    ('onehot', OneHotEncoder(sparse_output=False, handle_unknown='ignore'))
])

transformer = ColumnTransformer([
    ('num', num_pipeline, num_features),
    ('object', cat_pipeline, cat_features)
], remainder='passthrough')
```

```python
transformer.fit(X_train)
```

```python
X_train_transformer = pd.DataFrame(data=transformer.transform(X_train), columns=transformer.get_feature_names_out())
X_test_transformer = pd.DataFrame(data=transformer.transform(X_test), columns=transformer.get_feature_names_out())
```

```python
rfc = RandomForestClassifier(n_estimators=1)
```

```python
rfc.fit(X_train_transformer, y_train)
```

```python
rfc.score(X_train_transformer, y_train), rfc.score(X_test_transformer, y_test)
```

```python
from sklearn.metrics import accuracy_score, recall_score, precision_score, f1_score
```

```python
preds = rfc.predict(X_test_transformer)
```

```python
preds
```

```python
accuracy_score(y_test, preds)
```

```python
recall_score(y_test, preds, average='weighted')
```

```python
precision_score(y_test, preds, average='weighted')
```

```python
f1_score(y_test, preds, average='weighted')
```
