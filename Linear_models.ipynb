Mini Batch Gradient Descent

```python
np.random.permutation(m)
```

```python
from math import ceil

n_epochs = 50
eta = 0.1
m = len(X_b)
minibatch_size = 20     # 1 destede 20 dene olsun
n_batches_per_epoch = ceil(m / minibatch_size)   # ceil bolende yuxariya yuvarlaqladiririq

np.random.seed(42)
theta = np.random.randn(2, 1)

for epoch in range(n_epochs):
  shuffled_indices = np.random.permutation(m)
  X_b_shuffled = X_b[shuffled_indices]
  y_shuffled = y[shuffled_indices]
  for iteration in range(0, n_batches_per_epoch):
    idx = iteration * minibatch_size
    xi = X_b_shuffled[idx : idx + minibatch_size]
    yi = y_shuffled[idx : idx + minibatch_size]
    gradients = 2 / minibatch_size * xi.T @ (xi @ theta - yi)
    theta = theta - eta * gradients
```

```python
x = np.array([60, 80, 90, 100, 120])
y = np.array([50, 60, 70, 80, 90])

plt.scatter(x, y, color = 'red', marker = '+')
plt.show()
```

```python
def meyil_azaltma(x, y, L, iterasiya):
    m = c = 0
    n = len(y)

    for i in range(iterasiya):
        y_texmini = m * x + c
        itki_funksiyasi = (1/n) * sum((y - y_texmini) ** 2)
        md = -(2/n)*sum(x * (y - y_texmini))
        cd = -(2/n)*sum(y - y_texmini)

        m = m - L * md
        c = c - L * cd

    return m, c

m, c = meyil_azaltma(x, y, 0.0001, 100000)
print(m, c)
```
