```python
df = pd.read_csv('/content/train.csv')
df.head()
```

```python
df.drop(['ID', 'Customer_ID', 'Name', 'SSN', 'Type_of_Loan', 'Credit_History_Age'], axis=1, inplace=True)
```

```python
df.head()
```

```python
df['Age'] = pd.to_numeric(df['Age'].str.replace('_', ''))
```

```python
df['Age'] = np.where(df['Age'] < 18, None, df['Age'])
df['Age'] = pd.to_numeric(df['Age'])

df['Annual_Income'] = pd.to_numeric(df['Annual_Income'].str.replace('_', ''))

df['Num_of_Loan'] = pd.to_numeric(df['Num_of_Loan'].str.replace('_', ''))

df['Num_of_Delayed_Payment'] = pd.to_numeric(df['Num_of_Delayed_Payment'].str.replace('_', ''))

df['Changed_Credit_Limit'] = pd.to_numeric(df['Changed_Credit_Limit'].str.replace('_', ''))

df['Outstanding_Debt'] = pd.to_numeric(df['Outstanding_Debt'].str.replace('_', ''))

df['Amount_invested_monthly'] = pd.to_numeric(df['Amount_invested_monthly'].str.replace('_', ''))

df['Monthly_Balance'] = pd.to_numeric(df['Monthly_Balance'].str.replace('_', '').str.replace('-', '').str.replace('333333333333333333333333333', ''))
```

```python
X = df.drop('Credit_Score', axis=1)
y = df['Credit_Score'].copy()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
num_features = X_train.select_dtypes(include = [np.number]).columns
cat_features = X_train.select_dtypes(exclude = [np.number]).columns
```

```python
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.compose import ColumnTransformer
from sklearn.ensemble import RandomForestClassifier
```

```python
num_pipeline = Pipeline([
    ('impute', SimpleImputer(strategy='median')),
    ('scale', StandardScaler())
])

cat_pipeline = Pipeline([
    ('impute', SimpleImputer(strategy = 'most_frequent')),
    ('onehot', OneHotEncoder(handle_unknown='ignore', sparse_output=False))
])

transformer = ColumnTransformer([
    ('num', num_pipeline, num_features),
    ('cat', cat_pipeline, cat_features)
], remainder='passthrough')
```

```python
transformer
```

```python
model = RandomForestClassifier()
```

```python
full_pipeline = Pipeline([
    ('preprocessing', transformer),
    ('estimator', model)
])
```

```python
full_pipeline
```

```python
full_pipeline.fit(X_train, y_train)
```

```python
# print(f'Train score:{full_pipeline.score(X_train, y_train)}\nTest score: {full_pipeline.score(X_test, y_test)'}
print(f'Train score: {full_pipeline.score(X_train, y_train)}\nTest score: {full_pipeline.score(X_test, y_test)}')
```
