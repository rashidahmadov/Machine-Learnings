# **Tuple_Slice**

```python
tuple1 = (12, 32.3, 65, 2, 43, 86) #Created a tuple
print(tuple1)
print(tuple1[1:5]) #prints the 2nd element - 5th element
print(tuple1[:4]) #prints the 1st element - 4th element
print(tuple1[2:]) #prints the 3rd element - to the last element
print(tuple1[:]) #prints all the elements
print(tuple1[: : 2]) #prints 1st element - last element with a step of 2
```

```python
#Sort the tuple by the 2nd element of the tuple
tuple1 = (('Bob', 27),('Tina', 17),('Sam', 34), ('Chris', 22))

temp = []

newTuple = ()

for n, t in tuple1:
  temp.append(t)

temp.sort()

for i in temp:
  for x,y in tuple1:
    if i == y:
      newTuple += (x,y)

print(newTuple)
```
