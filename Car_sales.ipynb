```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
!wget https://raw.githubusercontent.com/Rstam59/TaskDataRepoForStudents/refs/heads/main/car-sales-extended-missing-data.csv
```

```python
df = pd.read_csv('/content/car-sales-extended-missing-data.csv')
df
```

```python
df.isna().sum()
```

```python
df.dropna(subset='Price', inplace=True)
```

```python
df
```

```python
for column in df.columns:
  if df[column].dtype == float or df[column].dtype == int:
    df[column] = df[column].fillna(df[column].median())
  else:
    df[column] = df[column].fillna(df[column].mode()[0])
```

```python
df.isna().sum()
```

```python
df
```

```python
X = df.drop('Price', axis=1)
y = df['Price'].copy()
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
X_train
```

```python
y_train
```

```python
from sklearn.preprocessing import OrdinalEncoder, OneHotEncoder
```

```python
oe = OrdinalEncoder()
```

```python
X_train_encoded = oe.fit_transform(X_train[['Make', 'Colour']])
X_test_encoded = oe.transform(X_test[['Make', 'Colour']])
```

```python
X_train_encoded
```

```python
X_train['Make'] = X_train_encoded[:, 0]
X_train['Colour'] = X_train_encoded[:, 1]
X_test['Make'] = X_test_encoded[:, 0]
X_test['Colour'] = X_test_encoded[:, 1]
```

```python
X_train
```

```python
X_test
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
rfr = RandomForestRegressor()
```

```python
rfr.fit(X_train, y_train)
```

```python
rfr.score(X_train, y_train)
```

```python
rfr.score(X_test, y_test)
```
